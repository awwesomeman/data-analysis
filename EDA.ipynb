* plot the categorical data (with time series)

```python
def order(x):
    list = []
    for j in range(x[1]):
        for i in range(x[0]):
            list.append((i,j))
    return list
```

```python
for n in  np.arange(6, cat.shape[1], 13):
    
    fig, axes = plt.subplots(7,2, figsize=(12, 10))
    for i,j in enumerate(cat.columns.values[n:n+13]):
        cat[j].value_counts().plot.barh(ax = axes[order([7,2])[i]])
        axes[order([7,2])[i]].set_title(j)
    plt.tight_layout()
    plt.show()

    print("\n")
    print("===============================================================================================================")
    print("\n")
    print("===============================================================================================================")
```

```python
for n in  np.arange(0, num.shape[1], 13):
    
    fig, axes = plt.subplots(7,2, figsize=(12, 10))
    for i,j in enumerate(num.columns.values[n:n+13]):
        data = num[j].values
        data = data[data!=0]
        axes[order([7,2])[i]].hist(data,bins= 60) #num[j].value_counts().plot.hist(ax = axes[order([7,2])[i]])
        axes[order([7,2])[i]].set_title(j)
    plt.tight_layout()
    plt.show()

    print("\n")
    print("===============================================================================================================")
    print("\n")
    print("===============================================================================================================")
```

```python
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 300)
sns.heatmap(num.corr())
```

```python
df2 = num.copy()
df2.columns = "c_" + df2.columns
df3 = pd.concat([num, df2], axis=1)

# get the correlation coefficient between the different columns
corr = df3.corr()
arr_corr = corr.values
# mask out the top triangle
arr_corr[np.triu_indices_from(arr_corr)] = np.nan

fig, ax = plt.subplots(figsize=(24, 18))

hm = sns.heatmap(arr_corr, cbar=True, vmin=-0.5, vmax=0.5,
                 fmt='.2f', annot_kws={'size': 3}, annot=True, 
                 square=True, cmap=plt.cm.Blues)

ticks = np.arange(corr.shape[0]) + 0.5
ax.set_xticks(ticks)
ax.set_xticklabels(corr.columns, rotation=90, fontsize=8)
ax.set_yticks(ticks)
ax.set_yticklabels(corr.index, rotation=360, fontsize=8)

ax.set_title('correlation matrix')
plt.tight_layout()
plt.savefig("corr_matrix_incl_anno_double.png", dpi=300)
```
